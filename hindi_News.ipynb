### GET Forms

```python
import requests
from bs4 import BeautifulSoup
```

```python
content = requests.get("https://www.amarujala.com/")
```

```python
doc = BeautifulSoup(content.text, 'html.parser')
```

```python
# name_tags = doc.find_all("div", {"class": "image-caption"})

name_tags=doc.select("div.image-caption > h3 > a")
```

```python
# Grab all of the titles
# name_tags = doc.find_all("div", {"class": "image-caption"})

# Let's print the first 5

# print(name_tags)
for name in name_tags:
    print(name.get_text())
```

```python
response = requests.get("https://search.dca.ca.gov/results")
doc = BeautifulSoup(response.text, 'html.parser')

# Grab all of the rows
name_tags = doc.find_all("h3")

# Let's print the first 5
for name in name_tags:
    print(name.text)
```
